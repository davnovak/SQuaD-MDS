```python
from sklearn.datasets import make_blobs
import ViVAE # for plotting; `pip install --upgrade git+https://github.com/saeyslab/ViVAE.git`
import SQuadMDS
```

```python
X,y = make_blobs(1000, 50)
```

```python
model = SQuadMDS.SQuadMDS(n_components=2, tsne_hybrid=True)
```

```python
emb = model.fit_transform(X)
```

```python
ViVAE.plot_embedding(emb, y)
```
